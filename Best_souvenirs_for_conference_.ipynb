```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 

gift = pd.read_csv("https://raw.githubusercontent.com/annsyj94/Short-Data-Analysis-/main/New%20souvenir%20for%20%20GA%20event%20%20-%20GA%20Souvenir%20.csv")
gift.head(5)
```

```python
gift.info()
```

```python
# Use "for" to see the list 

data = gift['gift ']
target = gift['count ']

for index,value in enumerate(data):
  print(index,value)
```

```python
# Test the dataset by barchart 
import numpy as np

x = np.arange(20)

sns.set_style("whitegrid")
plt.figure(figsize=(50,20))
plt.bar(x,target)
plt.xticks(x,data)
plt.show()
```

```python
# Create a pie chart with legend 
colors = sns.color_palette('Set2',len(data))

sns.set_style("ticks")
plt.figure(figsize = (15,20))
plt.pie(target, labels=data, autopct='%1.1f%%', startangle = 90, textprops={'fontsize':11}, colors = colors, labeldistance=1.1)
plt.legend(data,borderaxespad=0., bbox_to_anchor=(0,1.02))
plt.title("Best Souvenirs for Conference ", fontsize = 15)
plt.show()

# Save it as a picture 
plt.savefig("Best Souvenirs for Conference")
```

```python
# Create a pie chart without legend 
colors = sns.color_palette('Set2',len(data))

sns.set_style("ticks")
plt.figure(figsize = (15,20))
plt.pie(target, labels=data, autopct='%1.1f%%', startangle = 90, textprops={'fontsize':11}, colors = colors, labeldistance=1.1)
plt.title("Best Souvenirs for Conference ", fontsize = 15)
plt.show()

plt.savefig("Best Souvenirs for Conference without Legend")
```
